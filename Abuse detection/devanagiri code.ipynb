```python

stp=list(open(r"C:\Users\anshika srivastava\Desktop\abusestopword.txt"  ,'r' , encoding='utf-8'))#self made list of stopwords
stop=set(stopwords.words('english'))

for i in range(0 , len(stp)):
    stp[i]=re.sub('\n', '',stp[i])

corpus = []
 
data=df.text.tolist()
valueto=df.value.tolist()

#preprocessing the data
for i in range(0 ,len(data)):
    data[i] =re.sub('@[\w]+','',data[i])
    data[i] = re.sub(r'\/\/.*[\r\n]*', '', data[i], flags=re.MULTILINE)
    data[i] = re.sub('pic.*[\r\n]*', '', data[i])
    data[i] = re.sub('#', '' , data[i])
    data[i] = re.sub('\d+', '' , data[i]) 
    data[i] = re.sub('http:', '' , data[i])
    data[i] = re.sub('https:' , '' , data[i])
   
    data[i] = data[i].translate(data[i].maketrans('', '', '"'))
    data[i] = data[i].translate(data[i].maketrans('', '', string.punctuation))
for j in range(300 , len(data)):
    data[j]=data[j].lower()
    
#implementing stopwords
for i in range(0 ,len(data)):
    data[i] = word_tokenize(data[i])
    data[i] = ([word for word in data[i] if word  not in stop])
    data[i] = ([word for word in data[i] if word  not in stp])
    
    data[i] =' '.join(str(data[i]).split(' '))
    
    corpus.append(data[i])
dfres = pd.DataFrame({'text': data, 'value': valueto})
print(data[:200])
```

```python
#bow model
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report
from sklearn.pipeline import Pipeline
```

```python
import regex 

def custom_analyzer(text):
    words = regex.findall(r'\w{2,}', text) # extract words of at least 2 letters since hindi words were getting stemmed
    for w in words:
        yield w


cv = CountVectorizer(analyzer = custom_analyzer )
X = cv.fit_transform(corpus).toarray()
print(X.shape)
```

```python
y=pd.get_dummies(dfres['value'])
y=y.iloc[:,1].values
print(y)
y.shape
```

```python
#multinomial naive bayes in bow
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 0)
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)

y_pred=spam_detect_model.predict(X_test)

ac_mnb = accuracy_score(y_pred, y_test)
print('accuracy %s' % ac_mnb)


confusion_matrix(y_test , y_pred)
```

```python
#logistic regression on bow
from sklearn.linear_model import LogisticRegression
lgmodel=LogisticRegression().fit(X_train, y_train)
y_lr=lgmodel.predict(X_test)
ac=accuracy_score(y_lr , y_test)
print('accuracy %s'% ac)
confusion_matrix(y_test , y_lr)
```
